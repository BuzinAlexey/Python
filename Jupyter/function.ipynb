```python
# Дана функция f(x) = -12x^4*sin(cos(x)) - 18x^3+5x^2 + 10x - 30
#1. Определить корни
#2. Найти интервалы, на которых функция возрастает
#3. Найти интервалы, на которых функция убывает
#4. Построить график
#5. Вычислить вершину
#6.Определить промежутки, на котором f > 0
#7.Определить промежутки, на котором f < 0
from sympy import*
from sympy.plotting import plot
from matplotlib import*
init_printing()
x = Symbol('x')
f = -12*x**4*sin(cos(x)) - 18*x**3+5*x**2 + 10*x - 30
plot(f)
```

```python
print(f'горизонтальная асимптота слева: {limit(f, x, -oo) }')
print(f'горизонтальная асимптота справа: {limit(f, x, oo) }')
print(f'точка пересечения с осью Y: [0 ,{f.subs(x, 0)}]')
```
