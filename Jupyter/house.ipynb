# Задание
К аналитику обратился риэлтор c задачей выставить ценник для продажи объектов недвижимости площадью 48м2 и 54м2. В ходе работы над задачей аналитик выяснил, что:  

объект 31м2 стоит $19310

объект 51м2 стоит $52150

объект 61м2 стоит $74570

Вопрос: что. сказать риелтору?


# метод 1

Найдем стоимость 1м2 каждого объекта и найдем среднюю стоимость 1м2

```python
from sympy import*
from sympy.plotting import plot
from matplotlib import*
import matplotlib.pyplot as plt

init_printing()
object_1 = 19310
object_2 = 52150
object_3 = 74570
s1 = 31
s2 = 51
s3 = 61
print(f'цена 1 м2 объекта 1: {object_1/s1} $')
print(f'цена 1 м2 объекта 2: {object_2/s2} $')
print(f'цена 1 м2 объекта 3: {object_3/s3} $')
C = (object_1/s1+object_2/s2+object_3/s3)/3
print(f'Среджняя цена за 1 м2: {C}')
```

найдем цены нужных объектов


```python
s1_ob1 = 48
s2_ob2 = 54
print(f'цена объекта  = 48 м2: {C*s1_ob1} $')
print(f'цена объекта  = 54 м2: {C*s2_ob2} $')
```
